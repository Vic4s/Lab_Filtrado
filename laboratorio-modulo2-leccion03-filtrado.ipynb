#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
cond_1 = df_merge["type"] == "Movie"
cond_2 = df_merge["Ano"] == 2021
df_movie_2021 = df_merge[cond_1 & cond_2]
df_movie_2021
```

Los valores de la columna Años los trata como float:

```python
df_movie_2021.info()
```

Voy a convertirl los valores de esa columna a int

```python
# Si hubiera valores nulos:
# df_merge['Ano'].fillna(0).astype(int)

df_movie_2021["Ano"] = df_movie_2021[["Ano"]].astype(int)
df_movie_2021.info()
```

```python
df_movie_2021.head()
```

No me dejaba convertir la columna duration a int para poder operar con ella, así que visualizo qué tipos únicos tengo y veo que hay NaN:

```python
print(df_merge["duration"].unique())
```

Intentaba rellenar los NaN con 0 y luego convertirlos a string, pero no me funcionaba. Así que he hecho un reemplazo de la cadena "nan" a "0 min"

```python
df_merge['duration'] = df_merge['duration'].fillna("0 min")
df_merge["duration"] = df_merge["duration"].astype(str)
df_merge["duration"] = df_merge["duration"].replace("nan", "0 min")
```

```python
print(df_merge["duration"].unique())
```

```python
temporadas_series = []
for element in df_merge.duration:
    if "Season" in element:
        temporadas_series.append(element.split(" ")[0])
    else:
        temporadas_series.append(0)

print(temporadas_series)

df_merge["Temporadas_Series"] = temporadas_series
```

```python
duracion_peliculas = []
for element in df_merge.duration:
    if "min" in element:
        duracion_peliculas.append(element.split(" ")[0])
    else:
        duracion_peliculas.append(0)

print(duracion_peliculas)

df_merge["Duracion_peliculas"] = duracion_peliculas
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?
